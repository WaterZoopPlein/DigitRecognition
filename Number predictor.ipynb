```python
def sigmoid(z):
    '''Calculating sigmoid function, giving z'''
    return 1 / (1+np.e**(-z))

def predict(theta,input_x):
    
    m = np.shape(input_x)[0]
    n = np.shape(input_x)[1]
    
    input_x = np.append(1,input_x)
    theta = np.reshape(theta,(k,n+1))
    
    h = sigmoid(np.matmul(theta, np.transpose(input_x)))
    h = np.transpose(h)
    
    predicted = np.argmax(h)
    
    return predicted, h
```

```python
k = 10
theta = np.loadtxt('theta')

digitArr = np.array(img.imread('test.png'))
digitArr = digitArr[:,:,1]
digitArr = np.reshape(digitArr,(1,784))

pred, h = predict(theta,digitArr)
print("MY PREDICTION IS {}".format(pred))

%matplotlib notebook
fig, [ax1,ax2] = plt.subplots(1,2,figsize=(10, 5))

digitArr = np.reshape(digitArr,(28,28))

ax1.imshow(digitArr)

x_ticks = np.arange(10)
ax2.bar(x_ticks,height=h)
ax2.set_xticks(x_ticks)
ax2.set_ylabel('Certainty rate')
ax2.set_ylim(0,1.05)
fig.show()
```
